```python
import csv
import matplotlib.pyplot as plt
import os
# CSV 파일 경로
print("goguma1")
dir_path = './profile'

file_list = os.listdir(dir_path)

# 데이터를 저장할 리스트 초기화
iter = [[] for _ in range(len(file_list))]
FLOPS = [[] for _ in range(len(file_list))]


# CSV 파일 읽기
for i, file in enumerate(file_list):
    file_path = os.path.join(dir_path, file)
    with open(file_path, 'r') as f:
        reader = csv.DictReader(f, delimiter='\t')

        # 각 행의 데이터 추출
        for idx, row in enumerate(reader):
            iter[i].append(idx)
            row_parse = row['epoch,iter,flops,latency']
            row_parse = row_parse.split(',')
            flops = row_parse[2]
            latency = row_parse[3]

            if flops.split()[1] == 'T':
                FLOPS[i].append(float(flops.split()[0])/float(latency.split()[0]))
            elif flops[1] == 'G':
                FLOPS[i].append(float(flops.split()[0])/1000 /float(latency.split()[0]))
            else:
                print(f'wrong unit flops{flops[1]}' )

gpu_names = os.popen("nvidia-smi --query-gpu=name --format=csv,noheader").read().strip().split('\n')
print(gpu_names)
avg = [[] for _ in range(len(file_list))]
for i in range(len(file_list)):
    avg[i] = sum(FLOPS[i][10:])/len(FLOPS[i][10:])
    print(f'avg: gpu{i}, {avg[i]}')

print(f'error gpu0,1: {(avg[0]-avg[1])/avg[1] * 100}')
print(f'error gpu1,2: {(avg[1]-avg[2])/avg[2] * 100}')
print(f'error gpu2,3: {(avg[2]-avg[3])/avg[3] * 100}')
print(f'error gpu0,2: {(avg[0]-avg[2])/avg[2] * 100}')
print(f'error gpu0,3: {(avg[0]-avg[3])/avg[3] * 100}')
print(f'error gpu1,3: {(avg[1]-avg[3])/avg[3] * 100}')
# 그래프 그리기
fig, axs = plt.subplots(1, 1, figsize=(12, 8))

for i in range(len(file_list)):
    axs.plot(iter[i][10:], FLOPS[i][10:], label=f'GPU{i},{gpu_names[i]}')
    
axs.set_xlabel('iter')
axs.set_ylabel('TFLOPS')
axs.set_title('FLOPS over iters')
axs.legend(bbox_to_anchor=(1.02,1), loc='upper left')
plt.show()
```
